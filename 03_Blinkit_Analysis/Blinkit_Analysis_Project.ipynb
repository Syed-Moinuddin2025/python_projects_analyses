### **KPI's REQUIREMENT**

```python
# Total Sales: The overall revenue generated from all items sold
total_sales = df['Sales'].sum()

#Average Sales: The average revenue per sale.
average_sales = df['Sales'].mean()

#Number of Items: The total count of different items sold.
number_of_items = df['Item Identifier'].nunique()
no_of_items_sold =  df['Sales'].count()

#Average Rating: The average customer rating for items sold. 
average_rating = df['Rating'].mean()

# üîç Display results
print(f"Total Sales: ${total_sales:.2f}")
print(f"Average Sales: ${average_sales:.2f}")
print(f"Number of Items: {number_of_items}")
print(f"Number of Items: {no_of_items_sold}")
print(f"Average Rating: {average_rating:.2f}")
```

```python
# üéØ KPI Calculation Block

# 1Ô∏è‚É£ Total Sales Value
total_sales = df['Sales'].sum()

# 2Ô∏è‚É£ Average Sales Value
average_sales = df['Sales'].mean()

# 3Ô∏è‚É£ Total Items Sold (rows/transactions)
total_items_sold = df['Sales'].count()

# 4Ô∏è‚É£ Unique Items Sold (distinct products)
unique_items_sold = df['Item Identifier'].nunique()

# 5Ô∏è‚É£ Average Customer Rating
average_rating = df['Rating'].mean()

# üßæ Display Summary with Emojis
print("üìäüîç Blinkit Sales Summary Report üîçüìä\n")
print(f"üí∞ Total Sales Value        : ${total_sales:,.2f}")
print(f"üìà Average Sales Value      : ${average_sales:,.2f}")
print(f"üßæ Total Items Sold         : {total_items_sold}")
print(f"üì¶ Unique Items Sold        : {unique_items_sold}")
print(f"‚≠ê Average Customer Rating  : {average_rating:.2f} / 5.00")

```

### **Chart‚Äôs Requirements**

#### **Total Sales by Fat Content:**

```python
sales_by_fat = df.groupby('Item Fat Content') ['Sales' ].sum()
print("Total Sales By fat:",sales_by_fat)
```

```python
sales_by_fat = df.groupby('Item Fat Content') ['Sales' ].sum()

plt.pie(sales_by_fat, labels= sales_by_fat.index,
autopct = '%.1f%%',
startangle = 90)

plt.title('Sales by Fat Content')
plt.axis('equal')
plt.show()
```

#### **Total Sales by Item Type**

```python
sales_by_type = df.groupby('Item Type')['Sales'].sum().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
bars = plt.bar(sales_by_type.index, sales_by_type.values, color='skyblue', edgecolor='black')

plt.xticks(rotation=-90)
plt.xlabel('Item Type')
plt.ylabel('Total Sales')
plt.title(' Total Sales by Item Type')

# Add value labels on bars
for bar in bars:
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        bar.get_height(),
        f'{bar.get_height():,.0f}',
        ha='center', va='bottom', fontsize=8
    )

plt.tight_layout()
plt.show()
```

```python
grouped = df. groupby(['Outlet Location Type', 'Item Fat Content' ]) ['Sales' ]. sum() . unstack()
grouped = grouped[ ['Regular', 'Low Fat' ]]

ax = grouped. plot(kind='bar', figsize=(8, 5), title='Outlet Tier by Item Fat Content')
plt.xlabel('Outlet Location Tier' )
plt.ylabel('Total Sales')
plt.legend(title='Item Fat Content')
plt.tight_layout()
plt.show()
```

#### **Total Sales by Outlet Establishment**

```python
sales_by_year = df.groupby('Outlet Establishment Year')['Sales'].sum().sort_index()

plt.figure(figsize=(9, 5))
plt.plot(sales_by_year.index, sales_by_year.values, marker='o', linestyle='-', color='blue')

plt.xlabel('Outlet Establishment Year')
plt.ylabel('Total Sales')
plt.title(' Total Sales by Outlet Establishment Year')

# Annotate values on the line
for x, y in zip(sales_by_year.index, sales_by_year.values):
    plt.text(x, y, f'{y:,.0f}', ha='center', va='bottom', fontsize=8)

#plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

#### **Sales by Outlet Size:**

#### **Sales by Outlet Location:**

```python
# üìä Total Sales by Outlet Location Type (Vertical Bar Chart)

sales_by_location = df.groupby('Outlet Location Type')['Sales'].sum().reset_index()
sales_by_location = sales_by_location.sort_values('Sales', ascending=False)

plt.figure(figsize=(6,6))
bars = plt.bar(
    sales_by_location['Outlet Location Type'],
    sales_by_location['Sales'],
    color=['#AED6F1', '#F9C79F', '#A9DFBF']
)

# Add labels on top of each bar
for bar in bars:
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        bar.get_height() + 6000,  # Offset so text is above bar
        f"${bar.get_height():,.0f}",
        ha='center',  fontsize=10, fontweight='bold'
    )

plt.title(' Total Sales by Outlet Location Type', fontsize=12, weight='bold')
plt.xlabel('Outlet Location Type', fontsize=10)
plt.ylabel('Total Sales', fontsize=10)
plt.xticks(fontsize=9)
plt.yticks(fontsize=9)
plt.tight_layout()
plt.show()
```

#### **1. Sales by Item Type:**

```python
# üìä Step 1: Total sales by Item Type (sorted ascending for horizontal bar)
sales_by_type = df.groupby('Item Type')['Sales'].sum().sort_values()

# üé® Step 2: Create a color palette
colors = plt.cm.Set2(np.linspace(0, 1, len(sales_by_type)))

# üìà Step 3: Create horizontal bar chart
plt.figure(figsize=(14,6))
bars = plt.barh(sales_by_type.index, sales_by_type.values, color=colors)

# üñçÔ∏è Step 4: Add data labels next to each bar
for bar in bars:
    plt.text(
        bar.get_width() + 10,  # small offset to the right
        bar.get_y() + bar.get_height() / 2,
        f"${bar.get_width():,.0f}",
        va='center', fontsize=9
    )

# üè∑Ô∏è Step 5: Add title and labels
plt.title(' Total Sales by Item Type', fontsize=13, weight='bold')
plt.xlabel('Total Sales')
plt.ylabel('Item Type')
plt.tight_layout()

# ‚úÖ Final: Show the chart
plt.show()
```

#### **2. Average Rating by Outlet Type**

```python
# Chart: Vertical Bar Chart
avg_rating_by_outlet = df.groupby('Outlet Type')['Rating'].mean().sort_values()

# üé® Step 2: Custom bar colors (you can tweak these)
colors = ['#FF6F61', '#6B5B95', '#88B04B', '#F7CAC9', '#92A8D1', 
          '#955251', '#B565A7', '#009B77', '#DD4124', '#45B8AC']

# üìà Step 3: Create bar chart
plt.figure(figsize=(5, 5))
bars = plt.bar(avg_rating_by_outlet.index, avg_rating_by_outlet.values, color=colors[:len(avg_rating_by_outlet)])

def new_func(bars):
    for bar in bars:
        plt.text(
            bar.get_x() + bar.get_width() / 2,
            bar.get_height() + 0,  # slight vertical offset
            f"{bar.get_height():.2f}",
            ha='center', va='bottom', fontsize=9, fontweight='bold'
        )

new_func(bars)

plt.title('Average Rating by Outlet Type')
plt.xlabel('Outlet Type')   
plt.ylabel('Average Rating')
plt.xticks(rotation=45) 
plt.tight_layout()
plt.show()
```

#### **3. Sales by Fat Content (Donut)**

#### **4. Top 10 Best-Selling Items**

```python
# üìä Chart: Vertical Bar Chart 
# üìä Step 1: Top 10 items by total sales
top_items = df.groupby('Item Identifier')['Sales'].sum().nlargest(10)

# üé® Step 2: Custom bar colors (you can tweak these)
colors = ['#FF6F61', '#6B5B95', '#88B04B', '#F7CAC9', '#92A8D1', 
          '#955251', '#B565A7', '#009B77', '#DD4124', '#45B8AC']

# üìà Step 3: Create bar chart
plt.figure(figsize=(8, 5))
bars = plt.bar(top_items.index, top_items.values, color=colors)

# üñçÔ∏è Step 4: Add data labels on top of each bar
for bar in bars:
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        bar.get_height() + 20,  # slight vertical offset
        f"${bar.get_height():,.0f}",
        ha='center', va='bottom', fontsize=9, fontweight='bold'
    )

# üè∑Ô∏è Step 5: Add titles and labels
plt.title(' Top 10 Best-Selling Items', fontsize=12, weight='bold')
plt.xlabel('Item Identifier')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.tight_layout()

# ‚úÖ Final: Show the chart
plt.show()
```

#### **6. Item Weight Distribution**

```python
#Chart: Histogram
import numpy as np
import matplotlib.pyplot as plt

# Get histogram data manually
counts, bins = np.histogram(df['Item Weight'].dropna(), bins=15)

# Generate different colors (can use any colormap)
colors = plt.cm.Set3(np.linspace(0, 1, len(counts)))  # Set3, Pastel1, viridis, etc.

# Plot each bar manually
plt.figure(figsize=(8, 4))
for i in range(len(counts)):
    plt.bar(
        bins[i],
        counts[i],
        width=bins[i+1] - bins[i],
        color=colors[i],
        edgecolor='black',
        align='edge'
    )

plt.title('Distribution of Item Weight (Multicolor)', fontsize=12, weight='bold')
plt.xlabel('Item Weight')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```
